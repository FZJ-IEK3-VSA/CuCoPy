## **CuCoPy**: **Cu**rrency **Co**nversion for **Py**thon

### Initialize Class

```python
from cucopy.currency import Currency
cur = Currency()
```

### Convert Currency using Exchange Rate and Inflation

```python
value_converted = cur.convert_currency(value=100, base_year="2019", base_currency="USD", target_year="2025", target_currency="EUR", operation_order="inflation_first")
print(value_converted)
```

### Currency Exchange Rate Adjustment


```python
value_er = cur.get_exchanged_value(100, "2024", "EUR", "USD")
print(value_er)
```

### Currency Inflation Adjustment

#### Equivalent

```python
value_inf = cur.adjust_for_inflation(100, "EUR", "2020", "2025", mode="equivalent")
print(value_inf)
```

#### Purchasing Power

```python
value_inf = cur.adjust_for_inflation(100, "EUR", "2020", "2025", mode="purchasing_power")
print(value_inf)
```

### Manually add new notations

```python
cur.unique_notation("Q", 15)
print(cur._allowed_notations)
```

```python
value_er = cur.get_exchanged_value(100, "2020", "EUR", "USD", value_notation="Q")
print(value_er)
```
